### DATA VISUALIZATION WITH MATPLOTLIB 

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(df['Mes'], df['data science'], label='data science')
```

```python
plt.plot(df['Mes'], df['data science'], label='data science')
plt.plot(df['Mes'], df['machine learning'], label='machine learning')
plt.plot(df['Mes'], df['deep learning'], label='deep learning')
plt.xlabel('Dates')
plt.ylabel("popularity")
plt.title("popularity of AI terms by Dates")
plt.grid(True)
plt.legend()
plt.show()
```

### WITHOUT  GRID

```python
plt.plot(df['Mes'], df['data science'], label='data science')
plt.plot(df['Mes'], df['machine learning'], label='machine learning')
plt.plot(df['Mes'], df['deep learning'], label='deep learning')
plt.xlabel('Dates')
plt.ylabel("popularity")
plt.title("popularity of AI terms by Dates")
plt.legend()
plt.show()
```

```python
### We can draw the graph with different styles for the points of each variable: DATA SCIENCE

plt.plot(df['Mes'], df['data science'], 'r')
plt.plot(df['Mes'], df['data science']*2, 'g')
plt.plot(df['Mes'], df['data science']*3, 'b')
```

```python
### We can draw the graph with different styles for the points of each variable: DATA SCIENCE, MACHINE LEARNING, DEEP LEARNING

plt.plot(df['Mes'], df['data science'])
plt.plot(df['Mes'], df['machine learning'])
plt.plot(df['Mes'], df['deep learning'])
```

```python
### multiple graphics in one figure

fig, axes = plt.subplots(2,2)
axes[0,0].hist(df['data science'])
axes[0,1].plot(df['Mes'], df['data science'])
axes[1,0].plot(df['Mes'], df['machine learning'])
axes[1,1].plot(df['Mes'], df['deep learning'])
```

```python
### multiple graphics in one figure with scatter plot

fig, axes = plt.subplots(2,2)
axes[0,0].hist(df['data science'])
axes[0,1].scatter(df['machine learning'], df['data science'])
axes[1,0].scatter(df['deep learning'], df['machine learning'])
axes[1,1].scatter(df['data science'], df['deep learning'])
```

```python
plt.scatter(df['machine learning'], df['data science'])
```

```python
### multiple graphics in one figure with scatter plot

fig, axes = plt.subplots(2,2)
axes[0,0].hist(df['data science'])
axes[0,1].bar(df['machine learning'], df['data science'], width=0.8)
axes[1,0].bar(df['deep learning'], df['machine learning'])
axes[1,1].bar(df['data science'], df['deep learning'])
```
