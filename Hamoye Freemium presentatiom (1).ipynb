## IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## IMPORTING DATASETS WITH PANDAS 

```python
dataset = pd.read_csv("Placement_Data_Full_Class.csv")
dataset.head()
```

## RESULT FROM THE DATASET
#### We have Gender and Educational qualification data
#### We have all the educational performance(score) data

```python
dataset.fillna(value = 0, axis = 0, inplace = True)
```

## DATA VISUALIZATION ANALYSIS

```python
plt.figure(figsize = (15, 7))
plt.style.use('_classic_test_patch')

#Specialisation
plt.subplot(234)
ax=sns.countplot(x="specialisation", data=dataset,linewidth=5, 
                 edgecolor=sns.color_palette('flag', 3), facecolor=(0, 0, 0, 0))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)


#Gender
plt.subplot(231)
ax=sns.countplot(x="gender", data=dataset, linewidth=5,
                 edgecolor=sns.color_palette("flare_r", 3), facecolor=(0, 0, 0, 0))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)

#status of recriutment
plt.subplot(236)
ax=sns.countplot(x="status", data=dataset, linewidth=5,
                 edgecolor=sns.color_palette("hot", 3), facecolor=(0, 0, 0, 0))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)


#High secondary specialisation
plt.subplot(232)
ax=sns.countplot(x="hsc_s", data=dataset, linewidth=5,
                 edgecolor=sns.color_palette("gist_rainbow_r", 3), facecolor=(0, 0, 0, 0))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)


#Degree Type
plt.subplot(233)
ax=sns.countplot(x="degree_t", data=dataset, linewidth=5,
                 edgecolor=sns.color_palette("cool", 3), facecolor=(0, 0, 0, 0))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)


#Work Experience
plt.subplot(235)
ax=sns.countplot(x="workex", data=dataset, linewidth=5,
                 edgecolor=sns.color_palette("summer_r", 3), facecolor=(0, 0, 0, 0))
fig = plt.gcf()
fig.set_size_inches(10,10)
ax.set_xticklabels(ax.get_xticklabels(),fontsize=12)
```

## CONCLUSION 
#### We have more male candidates than female
#### Candidates who did commerce in their hsc and  Comm&Mgmt in their degree are more than candidates who did science in both hsc and degree
#### Candidates from Marketing and Finance dual specialization are high
#### Most candidates from this dataset don't have any work experience
#### Most candidates from this dataset got placed in a company

```python
### Distribution of all Data in percentage

plt.figure(figsize = (15, 7))
plt.style.use('seaborn-white')
plt.subplot(231)
sns.histplot(dataset['ssc_p'])
fig = plt.gcf()
fig.set_size_inches(12,12)

plt.subplot(232)
sns.histplot(dataset['hsc_p'])
fig = plt.gcf()
fig.set_size_inches(12,12)

plt.subplot(233)
sns.histplot(dataset['degree_p'])
fig = plt.gcf()
fig.set_size_inches(12,12)

plt.subplot(234)
sns.histplot(dataset['etest_p'])
fig = plt.gcf()
fig.set_size_inches(12,12)

plt.subplot(235)
sns.histplot(dataset['mba_p'])
fig = plt.gcf()
fig.set_size_inches(12,12)

plt.subplot(236)
sns.histplot(dataset['salary'])
fig = plt.gcf()
fig.set_size_inches(12,12)
```

## CONCLUSION
#### All the distributions follow normal distribution except salary feature
#### Most of the candidates educational performances are between 60-80%

```python
plt.style.use('seaborn-white')
f,ax=plt.subplots(1,2,figsize=(12,7))
dataset['workex'].value_counts().plot.pie(explode=[0,0.05],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Work_experience')
sns.countplot(x = 'workex',hue = "status",data = dataset)
ax[1].set_title('Impact of experience on placement')
plt.show()
```

## CONCLUSION 
#### -There are 66.6% of candidates who never had any work experience
#### -Candidates who never had work experience got hired more than those who had experience
#### -from the visualization above, we can conclude that work experience doesn't influence a candidate in the recruitment process
